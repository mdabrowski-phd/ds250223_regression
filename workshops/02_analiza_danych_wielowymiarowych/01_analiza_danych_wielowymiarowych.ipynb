#Rozkad wielomianowy
Zmienna losowa $(X_1, X_2, \ldots, X_k)$ ma rozkad wielomianowy, je偶eli 
\begin{equation*}
    P(X_1=n_1, \ldots, X_k=n_k)=\frac{n!}{n_1!\cdot\ldots\cdot n_k!}p_1^{n_1}\cdot\ldots\cdot p_k^{n_k},
\end{equation*}
gdzie $p_i\in(0,1)$, $i=1,2,\ldots, k$, $p_1+p_2+\ldots+p_k=1$, $n_1+n_2+\ldots+n_k=n$.

Rozkad wielomianowy jest uog贸lnieniem rozkadu dwumianowego i opisuje rozkad wynik贸w przy $n$-krotnym powt贸rzeniu dowiadczenia o $k$ mo偶liwych rezultatach. 

$X_i$ oznacza liczb wynik贸w $i$-tego typu w serii.

Rozkad wielomianowy mo偶na wykorzysta do obliczenia prawdopodobiestwa w sytuacjach, w kt贸rych istniej wicej ni偶 dwa mo偶liwe wyniki.

\begin{equation*}
    f(x, y) = \frac{1}{2\pi \sigma_X \sigma_Y\sqrt{1-\rho^2}} \exp\Bigg\{-\frac{1}{2(1-\rho^2)}\Bigg[\frac{(x-\mu_X)^2}{\sigma^2_X} - 2\rho\frac{(x-\mu_X)(y-\mu_Y)}{\sigma_X\sigma_Y} + \frac{(y-\mu_Y)^2}{\sigma^2_Y} \Bigg] \Bigg\}.

#Kowariancja
Kowariancj zmiennych losowych $X$ i $Y$ nazywamy liczb $\text{Cov}(X, Y)$ okrelon wzorem
\begin{equation*}
    \text{Cov}(X, Y) = \mathbb{E}[(X - \mathbb{E}X)(Y-\mathbb{E}Y) ].
\end{equation*}

Kowariancja mo偶e by traktowana jako pewna miara zgodnoci dw贸ch zmiennych losowych, bdc redni wartoci iloczynu odchyle obu zmiennych od ich wartoci oczekiwanych. 

Bezporednio z definicji wynika, 偶e 
\begin{equation*}
    \text{Cov}(X, Y) = \mathbb{E}(XY) - (\mathbb{E}X)(\mathbb{E}Y).
\end{equation*}

Gdy zmienna losowa $(X, Y)$ jest typu dyskretnego, to
    \begin{equation*}
        \mathbb{E}(XY) = \sum_i\sum_k x_iy_kP(X=x_i, Y=y_k).
    \end{equation*}
Gdy zmienna losowa $(X, Y)$ jest typu cigego, to
    \begin{equation*}
        \mathbb{E}(XY) = \int_{-\infty}^{+\infty}\int_{-\infty}^{+\infty} xyf(x,y)dxdy.
    \end{equation*}

Wasnoci:
    

*   Je偶eli $\text{Cov}(X, Y)=0$, to zmienne losowe $X$ i $Y$ nazywamy eskorelowanymi. Wynika std, 偶e zmienne losowe niezale偶ne s jednoczenie nieskorelowane, ale zmienne losowe nieskorelowane mog by zale偶ne.
*   $\text{Var}(X\pm Y) = \text{Var}(X) + \text{Var}(Y) \pm 2\text{Cov}(X, Y)$.
*   $\text{Cov}(aX+b, cY+d) = ac\text{Cov}(X, Y)$.


#Wsp贸czynnik korelacji
Wsp贸czynnikiem korelacji zmiennych losowych $X$ i $Y$ nazywamy liczb
\begin{equation*}
    \rho(X, Y) = \frac{\text{Cov}(X, Y)}{\sqrt{\text{Var}(X)}\sqrt{\text{Var}(Y)}}
\end{equation*}
Wasnoci:

*    $|\rho(X, Y)|\leq 1$.
*    Je偶eli zmienne losowe $X$ i $Y$ s niezale偶ne, to $\rho(X, Y)=0$.
*   Dla dowolnych liczb rzeczywistych $a, b, c, d$ zachodzi
    \begin{equation*}
        |\rho(aX+b, cY+d)|=|\rho(X, Y)|.
    \end{equation*}
*    Zmienne losowe $X$ i $Y$ s zale偶ne liniowo wtedy i tylko wtedy, gdy $|\rho(X, Y)|=1$

#Korelacja Pearsona
Chcemy przybli偶a $\rho(X, Y)$ za pomoc pr贸by. Empirycznym odpowiednikiem korelacji jest korelacja Pearsona dana wzorem:
\begin{equation*} 
\hat{\rho}(X_1,...,X_n,Y_1,..,Y_n) = \frac{\sum_{i=1}^n(X_i-\bar{X})(Y_i-\bar{Y})}{\sqrt{\sum_{i=1}^n(X_i-\bar{X})^2}\sqrt{\sum_{i=1}^n(Y_i-\bar{Y})^2}}
\end{equation*}

#Macierz Kowariancji

Niech $(X_1, X_2, \ldots, X_n)$ bdzie wektorem losowym (o skadowych cakowalnych z kwadratem), wtedy \textbf{macierz kowariancji} jest okrelona nastpujco:
\begin{equation*}
    \sum = 
\begin{bmatrix}
\sigma^2_1 & \sigma_{12} & \ldots & \sigma_{1n}\\
\sigma_{21} & \sigma^2_2 & \ldots & \sigma_{2n}\\
\vdots & \ldots & \ddots & \vdots \\
\sigma_{n1} & \sigma_{n2} & \ldots & \sigma^2_n
\end{bmatrix}
\end{equation*}
gdzie:
*    $\sigma_i^2$ jest wariancj zmiennej losowej $X_i$;
*    $\sigma_{ij}=\text{Cov}(X_i, X_j)$ jest kowariancj midzy zmiennymi losowym $X_i$ oraz $X_j$.

#Wielowymiarowy rozkad normalny
Wektor losowy $(X_1,\ldots, X_n)$ ma rozkad $N$-wymiarowy rozkad normalny z macierz kowariancji $\sum$ oraz wektorem rednich $\mu$, je偶eli jej gsto $f$ wyra偶a si wzorem
\begin{equation*}
    f(x_1, \ldots, x_n) = \frac{1}{\sqrt{(2\pi)^n|\sum|}} \exp\Bigg\{-\frac{1}{2}(x-\mu)^T\Sigma^{-1}(x-\mu) \Bigg\}.
\end{equation*}
